```python
import pandas as pd
import numpy as np
import pickle
```

```python
variables = ["price", "tri", "volume(VO)", "mtbv", "cap", "bid", "ask", "price(EUR)"]
column_name = pd.read_excel('HW2 data 1.xlsx', sheet_name= "price").columns
```

```python
var_dict=dict()
for item in variables:
    var_dict[item] = pd.read_excel('HW2 data.xlsx', sheet_name=item)
    var_dict[item].columns = column_name
    var_dict[item].columns.values[1] = 'date'
    start_date = pd.to_datetime("2018-09-01")
    end_date = pd.to_datetime("2024-10-23")
    var_dict[item] = var_dict[item][(var_dict[item]['date'] >= start_date) & (var_dict[item]['date'] <= end_date)].reset_index(drop=True)
```

```python
price_df = var_dict['price']
tri_df = var_dict['tri']
volume_VO_df = var_dict['volume(VO)']
mtbv_df = var_dict['mtbv']
cap_df = var_dict['cap']
bid_df = var_dict['bid']
ask_df = var_dict['ask']
price_EUR_df = var_dict['price(EUR)']
```

```python
analyst_df = price_df.iloc[:,:2].copy()
analyst_df_1 = pd.read_excel('analyst_daily_changes.xlsx')
analyst_df_1.columns.values[0] = 'date'
analyst_df = analyst_df.merge(analyst_df_1, on=['date'], how='left')
analyst_df.iloc[:,2:] = analyst_df.iloc[:,2:].fillna(0)
analyst_df.columns = price_df.columns
analyst_df
```

```python
myday = price_df['date'].dt.strftime("%d-%m-%Y")
myday
```

```python
volume_df = volume_VO_df.copy()
volume_VO_df_filtered = volume_VO_df.iloc[:,2:].fillna(0)
price_EUR_df_filtered = price_EUR_df.iloc[:,2:].fillna(0)
volume_df.iloc[:,2:] = price_EUR_df_filtered * price_EUR_df_filtered * 1000
volume_df.columns = price_df.columns
volume_df
```

```python
# (ask-bid) / ((ask+bid)/2)
spread = ask_df.copy()
ask_df_filtered = ask_df.iloc[:,2:]
bid_df_filtered = bid_df.iloc[:,2:]
spread.iloc[:,2:] = (ask_df_filtered - bid_df_filtered) / ((ask_df_filtered + bid_df_filtered)/2)
spread
```

```python
tcost = spread.copy()
tcost.iloc[:,2:] = spread.iloc[:,2:]/ 2 + 0.0001
```

```python
tcost
```

```python
T = myday.shape[0]
isactive = price_df.copy()
isactive.iloc[:,2:] = 0
```

```python
# The stock must have valid price and total return index for at least 90% of the past 252 business days
# First 252 days no data
price_df_filtered = (price_df.iloc[:,2:].notna().rolling(window=252).sum().fillna(0) >=  (252 * 0.9))
tri_df_filtered = (tri_df.iloc[:,2:].notna().rolling(window=252).sum().fillna(0) >=  (252 * 0.9))
isactive.iloc[:,2:] = (price_df_filtered & tri_df_filtered).copy()
```

```python
# true if not dead
dead_filtered = price_df.iloc[:,2:].diff().fillna(0)
ten_days = (dead_filtered.rolling(window=10).sum() != 10).astype(bool)
isactive.iloc[:,2:] = (isactive.iloc[:,2:] & ten_days).copy()
```

```python
# first 21 days no data
tri_df_filtered_2 = (tri_df.iloc[:,2:].notna().rolling(window=21).sum() >=  (21 * 0.9)).astype(bool)
volume_df_filtered = (volume_df.iloc[:,2:].notna().rolling(window=21).sum() >=  (21 * 0.9)).astype(bool)
isactive.iloc[:,2:] = isactive.iloc[:,2:] & tri_df_filtered_2 & volume_df_filtered
```

```python
mtbv_df_filtered = mtbv_df.iloc[:,2:].notna()
cap_df_filtered = cap_df.iloc[:,2:].notna()
isactive.iloc[:,2:] = isactive.iloc[:,2:] & mtbv_df_filtered & cap_df_filtered
```

```python
# assume 21 days in one month
spread_filtered = (spread.iloc[:,2:].notna().rolling(window=21).sum() == 21)
isactive.iloc[:,2:] = isactive.iloc[:,2:] & spread_filtered
```

```python
# analyst recommendations
analyst_df_filtered = (analyst_df.iloc[:,2:] != 0).cumsum() != 0
isactive.iloc[:,2:] = isactive.iloc[:,2:] & analyst_df_filtered
```

```python
month_begin = pd.DataFrame(isactive['date'])
match_index = month_begin.groupby([month_begin['date'].dt.year, month_begin['date'].dt.month]).head(1).index
```

```python
isactive_filtered_1 = ((spread.iloc[:,2:] == False).rolling(window=21).sum() == 0) & (spread.iloc[:,2:] < 0.01)
isactive_filtered_2 = ((spread.iloc[:,2:] == False).rolling(window=21).sum() != 0) & (spread.iloc[:,2:] < 0.008)
```

```python
# average daily volume over the past month was more than 1 million euros per day
isactive_filtered_3 = ((spread.iloc[:,2:] == False).rolling(window=21).sum() == 0) & (volume_df.iloc[:,2:].rolling(window=21).mean() > 1000000)
isactive_filtered_4 = ((spread.iloc[:,2:] == False).rolling(window=21).sum() != 0) & (volume_df.iloc[:,2:].rolling(window=21).mean() > 1200000)
```

```python
isactive_filtered = isactive.copy()
isactive_filtered.iloc[match_index,2:] = isactive_filtered.iloc[match_index,2:] & (isactive_filtered_1.loc[match_index] | isactive_filtered_2.loc[match_index])
isactive_filtered.iloc[match_index,2:] = isactive_filtered.iloc[match_index,2:] & (isactive_filtered_3.loc[match_index] | isactive_filtered_4.loc[match_index])
isactive_filtered
```

```python
isactive1 = isactive.copy()
isactive1  = isactive1.replace(True, np.nan)
isactive1.loc[match_index] = isactive_filtered.loc[match_index]
isactive1 = isactive1.ffill()
isactive1
```

```python
price = price_df.iloc[:,1:].copy().set_index('date', drop=True) 
tri = tri_df.iloc[:,1:].copy().set_index('date', drop=True)
volume = volume_df.iloc[:,1:].copy().set_index('date', drop=True)
mtbv = mtbv_df.iloc[:,1:].copy().set_index('date', drop=True)
rec = analyst_df.iloc[:,1:].copy().set_index('date', drop=True)
tcost = tcost.iloc[:,1:].set_index('date', drop=True)
isactivenow = isactive1.iloc[:,1:].copy().set_index('date', drop=True)
```

```python
with open('hw2_SA.pkl', 'wb') as file:
    pickle.dump({'myday': myday, 
                'price': price, 
                'tri': tri, 
                'volume': volume,
                'mtbv': mtbv,
                'rec': rec,
                'tcost': tcost,
                'isactivenow': isactivenow
                }, file)
```
